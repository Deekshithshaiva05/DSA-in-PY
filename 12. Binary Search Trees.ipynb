# Search in a Binary Search Tree

https://takeuforward.org/data-structure/search-in-a-binary-search-tree-2

https://www.youtube.com/watch?v=KcNt6v_56cc

https://leetcode.com/problems/search-in-a-binary-search-tree/description/

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        
class Solution:
    def searchBST(self, root, target):
        
        while root and root.val != target:
            if target < root.val:
                root = root.left
            else:
                root = root.right            
        return root
    
root = TreeNode(4)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(1)
root.left.right = TreeNode(3)

obj = Solution()
result = obj.searchBST(root, 2)

if result:
    print("Node found:", result.val)
else:
    print("Node not found")
```

Time Complexity: O(log N),Each step eliminates half of the tree, just like binary search. However, in the worst case (unbalanced tree), it could be O(N).

Space Complexity:O(1),Iterative solution uses constant space as no recursion stack is involved.

# Find Min/Max in BST



```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


class Solution:

    def findMin(self, root):

        if root is None:
            return None

        while root.left:
            root = root.left

        return root.val

    def findMax(self, root):

        if root is None:
            return None

        while root.right:
            root = root.right

        return root.val


# 🔹 Driver Code
if __name__ == "__main__":

    root = TreeNode(8)
    root.left = TreeNode(3)
    root.right = TreeNode(10)

    root.left.left = TreeNode(1)
    root.left.right = TreeNode(6)

    root.right.right = TreeNode(14)

    solution = Solution()

    print("Minimum value in BST:", solution.findMin(root))
    print("Maximum value in BST:", solution.findMax(root))
```

# Floor and Ceil in a BST

https://www.youtube.com/watch?v=xm_W1ub-K-w&list=PLgUwDviBIf0q8Hkd7bK2Bpryj2xVJk8Vk&index=44
    


```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


class Solution:

    # Find Floor
    def findFloor(self, root, x):

        floor = -1

        while root:

            if root.val == x:
                return root.val

            if root.val < x:
                floor = root.val
                root = root.right

            else:
                root = root.left

        return floor

    # Find Ceil
    def findCeil(self, root, x):

        ceil = -1

        while root:

            if root.val == x:
                return root.val

            if root.val > x:
                ceil = root.val
                root = root.left

            else:
                root = root.right

        return ceil

if __name__ == "__main__":

    root = TreeNode(8)

    root.left = TreeNode(4)
    root.right = TreeNode(12)

    root.left.left = TreeNode(2)
    root.left.right = TreeNode(6)

    root.right.left = TreeNode(10)
    root.right.right = TreeNode(14)

    x = 5

    solution = Solution()

    print("Floor:", solution.findFloor(root, x))
    print("Ceil:", solution.findCeil(root, x))
```

# Insert a given node in BST

https://leetcode.com/problems/insert-into-a-binary-search-tree/description/

https://www.youtube.com/watch?v=FiFiNvM29ps

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


class Solution:

    def insertIntoBST(self, root, val):

        if root is None:
            return TreeNode(val)

        if val < root.val:
            root.left = self.insertIntoBST(root.left, val)

        else:
            root.right = self.insertIntoBST(root.right, val)

        return root


# 🔹 Inorder Traversal
def inorder(root):

    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

if __name__ == "__main__":

    root = TreeNode(8)

    root.left = TreeNode(4)
    root.right = TreeNode(12)

    solution = Solution()

    root = solution.insertIntoBST(root, 5)

    print("BST after insertion:")

    inorder(root)
```

# Delete a node in BST

https://leetcode.com/problems/delete-node-in-a-bst/description/

https://www.youtube.com/watch?v=kouxiP_H5WE

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        
class Solution:
    def deleteNode(self, root, key):
        
        if root is None:
            return None
        
        if key < root.val:
            root.left = self.deleteNode(root.left, key)
        elif key > root.val:
            root.right = self.deleteNode(root.right, key)
        else:
            # case 1 & 2
            
            # No left child
            if root.left is None:
                return root.right
            
            # No right child
            if root.right is None:
                return root.left
            
            # case 3: Two children
            successor = self.findMin(root.right)
            
            root.val = successor.val
            
            root.right = self.deleteNode(root.right, successor.val)
            
        return root
    
    def findMin(self, node):
        
        while node.left:
            node = node.left
            
        return node
    
def inorder(root):

    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)
        
if __name__ == "__main__":

    root = TreeNode(5)

    root.left = TreeNode(3)
    root.right = TreeNode(7)

    root.right.left = TreeNode(6)
    root.right.right = TreeNode(8)

    solution = Solution()

    root = solution.deleteNode(root, 7)

    print("BST after deletion:")

    inorder(root)
```

# Kth largest/smallest element in Binary Search Tree

https://takeuforward.org/data-structure/kth-largest-smallest-element-in-binary-search-tree
    


```python
# Bruth Force

class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.data = val
        self.left = left
        self.right = right

class Solution:
    def kLargesSmall(self, root, k):
        
        def inorder(node, values):
            
            if node:
                inorder(node.left, values)
                values.append(node.data)
                inorder(node.right, values)
                
        values = []
        
        inorder(root, values)
        
        kth_smallest =  values[k-1]
        kth_largest = values[-k]
        
        return [kth_smallest, kth_largest]

root = TreeNode(3)
root.left = TreeNode(1)
root.left.right = TreeNode(2)
root.right = TreeNode(4)

solution = Solution()
k = 1
print(solution.kLargesSmall(root, k))  
```

```python
# Optimal approach

class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.data = val
        self.left = left
        self.right = right

class Solution:
    def kthSmallest(self, root, k):
        self.k = k
        self.result = None
        self.inorder(root)
        return self.result
    
    def inorder(self, node):
        
        if node is not None:
            self.inorder(node.left)
            self.k -= 1
            if self.k == 0:
                self.result = node.data
                return
            self.inorder(node.right)
            
    def kthLargest(self, root, k):
        self.k = k
        self.result = None
        self.reverse_inorder(root)
        return self.result
    
    def reverse_inorder(self, node):
        
        if node is not None:
            self.reverse_inorder(node.right)
            self.k -= 1
            if self.k == 0:
                self.result = node.data
                return
            self.reverse_inorder(node.left)
    
    def kLargesSmall(self, root, k):
        return [self.kthSmallest(root, k), self.kthLargest(root, k)]
    
if __name__ == "__main__":
    root = TreeNode(3)
    root.left = TreeNode(1)
    root.left.right = TreeNode(2)
    root.right = TreeNode(4)
    
    solution = Solution()
    k = 1
    result = solution.kLargesSmall(root, k)
    
    print(result)
```

# Check if a tree is a BST or not

https://leetcode.com/problems/validate-binary-search-tree/description/

https://www.youtube.com/watch?v=f-sj7I5oXEI

```python
from typing import Optional

# Definition for a binary tree node
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:

        def validate(node, low, high):

            if not node:
                return True

            if not (low < node.val < high):
                return False

            return (
                validate(node.left, low, node.val) and
                validate(node.right, node.val, high)
            )

        return validate(root, float('-inf'), float('inf'))


# 🔹 Driver Code
if __name__ == "__main__":

    # Valid BST
    root = TreeNode(5)

    root.left = TreeNode(3)
    root.right = TreeNode(7)

    root.left.left = TreeNode(2)
    root.left.right = TreeNode(4)

    root.right.left = TreeNode(6)
    root.right.right = TreeNode(8)

    solution = Solution()

    if solution.isValidBST(root):
        print("Valid BST")
    else:
        print("Invalid BST")
```

# LCA in Binary Search Tree  -- (Lowest Common Ancestor (LCA))

https://www.youtube.com/watch?v=cX_kPV_foZc
    
https://leetcode.com/problems/lowest-common-ancestor-of-a-binary-search-tree/description/

```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None


class Solution:
    def lowestCommonAncestor(self, root, p, q):

        if root is None or root == p or root == q:
            return root

        # Search left subtree
        left = self.lowestCommonAncestor(root.left, p, q)

        # Search right subtree
        right = self.lowestCommonAncestor(root.right, p, q)

        # If p and q found in different subtrees
        if left and right:
            return root

        # Otherwise return non-null value
        return left if left else right

if __name__ == "__main__":

    root = TreeNode(3)

    root.left = TreeNode(5)
    root.right = TreeNode(1)

    root.left.left = TreeNode(6)
    root.left.right = TreeNode(2)

    root.right.left = TreeNode(0)
    root.right.right = TreeNode(8)

    solution = Solution()

    p = root.left        
    q = root.right       

    lca = solution.lowestCommonAncestor(root, p, q)

    print("Lowest Common Ancestor:", lca.val)
```

```python
# Recursive - optimize

class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None


class Solution:
    def lowestCommonAncestor(self, root, p, q):

        if p.val < root.val and q.val < root.val:
            return self.lowestCommonAncestor(root.left, p, q)

        if p.val > root.val and q.val > root.val:
            return self.lowestCommonAncestor(root.right, p, q)

        return root


# 🔹 Driver Code
if __name__ == "__main__":

    root = TreeNode(6)

    root.left = TreeNode(2)
    root.right = TreeNode(8)

    root.left.left = TreeNode(0)
    root.left.right = TreeNode(4)

    root.left.right.left = TreeNode(3)
    root.left.right.right = TreeNode(5)

    root.right.left = TreeNode(7)
    root.right.right = TreeNode(9)

    solution = Solution()

    p = root.left              # 2
    q = root.right             # 8

    lca = solution.lowestCommonAncestor(root, p, q)

    print("LCA:", lca.val)
```

# Construct a BST from a preorder traversal

https://www.youtube.com/watch?v=UmJT3j26t1I

https://leetcode.com/problems/construct-binary-search-tree-from-preorder-traversal/description/

```python
# Optimal Recursive Solution

from typing import List, Optional

class TreeNode:
    def __init__(self, val=0):
        self.val = val
        self.left = None
        self.right = None


class Solution:

    def bstFromPreorder(self, preorder: List[int]) -> Optional[TreeNode]:

        self.i = 0

        def build(bound=float('inf')):

            # Stop if all elements processed
            # OR current value exceeds bound
            if self.i == len(preorder) or preorder[self.i] > bound:
                return None

            # Create root node
            root = TreeNode(preorder[self.i])

            self.i += 1

            # Build left subtree
            root.left = build(root.val)

            # Build right subtree
            root.right = build(bound)

            return root

        return build()

def inorder(root):

    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

if __name__ == "__main__":

    preorder = [8, 5, 1, 7, 10, 12]

    solution = Solution()

    root = solution.bstFromPreorder(preorder)

    print("Inorder Traversal of Constructed BST:")

    inorder(root)
```

# Inorder Successor/Predecessor in BST

https://www.youtube.com/watch?v=SXKAD2svfmI

https://takeuforward.org/data-structure/inorder-successorpredecessor-in-bst



```python
# Brute Force

class TreeNode:
    def __init__(self, val=0):
        self.val = val
        self.left = None
        self.right = None
        
class Solution:
    def inorderSuccessor(self, root, p):
        
        # container for inorder traversal
        inorder : list[int] = []
        
        self.inorderTraversal(root, inorder)
        
        # locate index of p.val (or insertion index)
        idx = self.binarySearch(inorder, p.val)
        
        # handle edge cases: last element or invalid
        if idx == len(inorder) - 1 or idx == -1:
            return None
        
        return TreeNode(inorder[idx + 1])
    
    def inorderTraversal(self, root, inorder):
        
        if root:
            self.inorderTraversal(root.left, inorder)
            inorder.append(root.val)
            self.inorderTraversal(root.right, inorder)
            
    def binarySearch(self, arr, target):
        
        left, right = 0, len(arr) - 1
        
        while left <= right:
            mid = left + (right - left) // 2
            
            if arr[mid] == target:
                return mid
            elif arr[mid] < target:
                left = mid + 1
            else:
                right =  mid - 1
        return -1 if left == len(arr) else left
    
def print_in_order(root):
    
    if root is None:
        return
    
    print_in_order(root.left)
    print(root.val, end = " ")
    print_in_order(root.right)
    
def main():
    root = TreeNode(5)
    root.left = TreeNode(3)
    root.right = TreeNode(6)
    root.left.left = TreeNode(2)
    root.left.right = TreeNode(4)
    root.right.right = TreeNode(7)

    print("BST: ", end="")
    print_in_order(root)
    print()

    p = root.left.right

    solution = Solution()
    successor = solution.inorderSuccessor(root, p)

    if successor is not None:
        print(f"Inorder Successor of {p.val} is: {successor.val}")
    else:
        print(f"Inorder Successor of {p.val} does not exist.")

if __name__ == "__main__":
    main()
```

Time Complexity: O(N + logN) where N is the number of nodes of the binary search tree. O(N) to traverse all nodes and store them in an inorder array and O(log N) for the binary search.

Space Complexity: O(N) as an array of size N is used to store the inorder traversal of the binary search tree.

Space Complexity: O(1) as no additional data structure or memory allocation is done during the traversal and algorithm. 

```python
# Optimal Approach

class TreeNode:
    def __init__(self, x: int):
        self.val = x
        self.left = None
        self.right = None


class Solution:
    def inorderSuccessor(self, root: 'TreeNode', p: 'TreeNode') -> 'TreeNode | None':

        successor = None

        while root is not None:

            if p.val >= root.val:
                root = root.right

            else:
                successor = root
                root = root.left

        return successor


def print_in_order(root: 'TreeNode | None') -> None:

    if root is None:
        return

    print_in_order(root.left)
    print(root.val, end=" ")
    print_in_order(root.right)


def main():

    root = TreeNode(5)

    root.left = TreeNode(3)
    root.right = TreeNode(6)

    root.left.left = TreeNode(2)
    root.left.right = TreeNode(4)

    root.right.right = TreeNode(7)

    print("BST: ", end="")
    print_in_order(root)
    print()

    p = root.left.right

    solution = Solution()

    successor = solution.inorderSuccessor(root, p)

    if successor is not None:
        print(f"Inorder Successor of {p.val} is: {successor.val}")
    else:
        print(f"Inorder Successor of {p.val} does not exist.")


if __name__ == "__main__":
    main()
```

Time Complexity: O(H) where H is the height of the binary search tree as we are traversing along the height of the tree.

# Merge 2 BSTs

https://takeuforward.org/data-structure/bst-iterator

https://www.youtube.com/watch?v=D2jMcmxU4bs

```python
# Bruth Force

class Node:
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None

class Solution:
    # Function to traverse a BST and store elements in list
    def traverse(self, root, elements):
        
        if not root:
            return
        
        self.traverse(root.left, elements)
        elements.append(root.data)
        self.traverse(root.right, elements)

    def mergeBSTs(self, root1, root2):
        
        elements = []

        self.traverse(root1, elements)
        self.traverse(root2, elements)

        elements.sort()

        return elements

# Driver code
if __name__ == "__main__":
    
    root1 = Node(2)
    root1.left = Node(1)
    root1.right = Node(4)

    root2 = Node(3)
    root2.left = Node(0)
    root2.right = Node(5)

    sol = Solution()
    result = sol.mergeBSTs(root1, root2)

    print(*result)
```

Time Complexity: O((n+m)*log(n+m)), we traverse both the BSTs and sort all the elements.

Space Complexity: O(m+n), additional space required for storing elements of the two BSTs.

```python
# Optimal Opproach

class Node:
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None

class Solution:
    # Function to perform inorder traversal
    def inorderTraversal(self, root, arr):
        
        if not root:
            return
        
        self.inorderTraversal(root.left, arr)
        arr.append(root.data)
        self.inorderTraversal(root.right, arr)

    def mergeArrays(self, arr1, arr2):

        merged = []
        i = j = 0
        
        while i < len(arr1) and j < len(arr2):
            if arr1[i] < arr2[j]:
                merged.append(arr1[i])
                i += 1
            else:
                merged.append(arr2[j])
                j += 1

        merged.extend(arr1[i:])
        merged.extend(arr2[j:])
        return merged

    def mergeBSTs(self, root1, root2):

        arr1, arr2 = [], []
        
        self.inorderTraversal(root1, arr1)
        self.inorderTraversal(root2, arr2)
        
        return self.mergeArrays(arr1, arr2)

if __name__ == "__main__":
    
    root1 = Node(3)
    root1.left = Node(1)
    root1.right = Node(5)

    root2 = Node(4)
    root2.left = Node(2)
    root2.right = Node(6)

    sol = Solution()
    result = sol.mergeBSTs(root1, root2)

    print(*result)
```

Time Complexity: O(n+m), we traverse both the BSTs and merge two sorted lists.

Space Complexity: O(m+n), additional space required for storing elements of the two BSTs.
