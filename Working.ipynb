```python
# Solution 01: Naive Approach(Brute-force approach): 

def getSingleElement(arr):
    n = len(arr)
    for i in range(n):
        num = arr[i]
        cnt = 0
        
        for j in range(n):
            if arr[j] == num:
                cnt += 1
        if cnt == 1:
            return num

def main():
    arr = [4, 1, 2, 1, 2]
    ans = getSingleElement(arr)
    print("The single element is:", ans)


if __name__ == '__main__':
    main()
    
# Time Complexity: O(N2)
# Space Complexity : O(1)
```
