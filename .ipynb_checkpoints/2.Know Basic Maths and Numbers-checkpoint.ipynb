# 1. Count digits in a number

```python
# Brute Force Approach

def countdigits(n):
    count = 0
    while n > 0:
        count += 1
        n = n // 10
    return count

if __name__ == "__main__":
    n = 4567
    digits = countdigits(n)
    print(f"In {n} there are {digits} digits")
    
    
# Time Complexity: O(log10N + 1)
# Space Complexity : O(1)
```

```python
# Optimal Approach

import math

def countdigits(n):
    count = int(math.log10(n)+1)
    return count

if __name__ == "__main__":
    n = 45675
    digits = countdigits(n)
    print(f"In {n} there are {digits} digits")
    
# Time Complexity: O(1)
# Space Complexity : O(1)
```

# 2.Reverse Digits of A Number

```python
def revdig(n):
    rev_digits = 0
    while n > 0:
        last_digits = n % 10
        rev_digits = (rev_digits*10)+last_digits
        n = n // 10
    return rev_digits

if __name__ == "__main__":
    n = 123456789
    rd = revdig(n)
    print(f"the number {n} reverse is {rd}")
    
    
# Time Complexity: O(log10N + 1)
# Space Complexity : O(1)
```

```python
def revdig2(n):
    while n % 10 == 0 and n!= 0:
        n = n // 10
    rev_digit = 0
    while n > 0:
        last_digit = n % 10
        rev_digit = (rev_digit*10)+last_digit
        n = n // 10
    return rev_digit

if __name__ == "__main__":
    n = 123456789
    rd = revdig2(n)
    print(f"the number {n} reverse is {rd}")
    
# Time Complexity: O(log10N)
# Space Complexity : O(1)
```

# 3.Check if a number is Palindrome or Not

```python
def is_palindrome(n):
    rev_num = 0
    duplicate = n
    
    while n > 0:
        last_digits = n % 10
        rev_num = (rev_num*10)+last_digits
        n = n // 10
    if duplicate == rev_num:
        return True
    else:
        return False
    
if __name__ == "__main__":
    n = 123454321
    print(is_palindrome(n))
    
    
# Time Complexity: O(log10N+1)
# Space Complexity : O(1)
```

# Find all Palindrome Numbers in a given range

Problem Statement: Given a range of numbers, find all the palindrome numbers in the range.

Note: A palindromic number is a number that remains the same when its digits are reversed. OR & a palindrome is a number that reads the same forward and backward Eg: 121,1221, 2552

```python
mini = int(input())
maxi = int(input())

def is_palindrome(n):
    rev_num = 0
    duplicate = n
    
    while n > 0:
        last_digits = n % 10
        rev_num = (rev_num*10)+last_digits
        n = n // 10
    return duplicate == rev_num
    
arr =[]
for i in range(mini,maxi):
    if is_palindrome(i):
        arr.append(i)
        
print(*arr)
```

# Check if a number is Armstrong Number or not

```python
n = int(input())
original = n

# Step 1: Count digits
cnt = 0
temp = n
while temp > 0:
    cnt += 1
    temp = temp // 10

# Step 2: Calculate Armstrong sum
s = 0
while n > 0:
    digit = n % 10
    s += digit ** cnt
    n =n // 10

# Step 3: Compare
if s == original:
    print("True")
else:
    print("False")
```

```python
# Define a class to check Armstrong numbers

def is_armstrong(num):
    k = len(str(num))  # Number of digits
    sum = 0
    n = num

    while n > 0:
        ld = n % 10             # Last digit
        sum += ld ** k          # Add ld^k
        n = n // 10             # Remove digit

    return sum == num

# Test the class method
number = 153

if is_armstrong(number):
    print(f"{number} is an Armstrong number.")
else:
    print(f"{number} is not an Armstrong number.")
```

# Check whether a number is Perfect Number or not

```python
def is_perfect(n):
    summ = 0
    
    # Loop to find all divisors of the number and calculate their sum
    for i in range(1,n):
        if n % i == 0:
            summ += i # Add i to the sum of divisors
            
    # If the sum of divisors is equal to n, it is a perfect number
    return summ == n

n = 6

if is_perfect(n):
    print("True")
else:
    print("False")
```

# Maximum and Minimum Digit in a Number

```python
n = int(input())

mini = 0
maxi = 0

while n > 0:
    digit = n % 10
    if digit > maxi:
        maxi = digit
        
    if digit < mini:
        mini = digit
    n = n // 10

print("Maximum : ", maxi)
print("Minimum : ", mini)
```

# Print Fibonacci Series up to Nth term

```python
n = int(input())

if n == 0:
    print(0)
elif n == 1:
    print(0,1)
else:
    fib = [0]*(n+1)
    fib[0] = 0
    fib[1] = 1
    
    for i in range(2, n+1):
        fib[i] = fib[i-1] + fib[i-2]
        
    print(" ".join(str(num) for num in fib))
```

# Check if a number is a Strong Number or not

```python
n = int(input())

def factorial(x):
    if x < 1:
        return 1
    return x * factorial(x-1)

summ = 0
for _ in range(len(str(n))):
    ld = n % 10
    summ += factorial(ld)
    n = n // 10
print(summ)
```

# Find GCD of two numbers

```python
# Method 01: Brute Force

N1 = int(input())
N2 = int(input())

gcd = 1

for i in range(1, min(N1,N2)+1):
    if N1 % i == 0 and N2 % i == 0:
        gcd = i
print(gcd)
```

```python
# Method 2: By using Euclidean Algorithm

N1 = int(input())
N2 = int(input())

while N1 > 0 and N2 > 0:
    if N1 > N2:
        N1 = N1 % N2
    else:
        N2 = N2 % N1
        
if N1 == N2:
    print(N2)
    
print(N1)
```

# Program to Add two fractions

```python
class Solution:
    def gcd(self, a, b):
        if b == 0:
            return a
        return self.gcd(b, a % b)
    
    def addFraction(self, num1, den1, num2, den2):
        
        lcm = (den1 * den2) // self.gcd(den1, den2)
        
        newNum1 = num1 * (lcm // den1)
        newNum2 = num2 * (lcm // den2)
        
        resultNum = newNum1 + newNum2
        resultDen = lcm
        
        common = self.gcd(resultNum, resultDen)
        resultNum //= common
        resultDen //= common
        
        print(f"{resultNum}/{resultDen}")
        
sol = Solution()
sol.addFraction(1, 3, 1, 6)
```

# Replace all the 0’s with 1 in a given integer

```python
def replace1(s):
    if s == 0:
        return 1 
    
    result = 0
    place = 1
    
    while s > 0:
        ld = s % 10
        if ld == 0:
            ld = 1
        result = result + (ld * place)
        place = place * 10
        s = s // 10
        
    return result

s = 102003
print(replace1(s))
```

# Express given number as Sum of Two Prime Numbers

```python
class Solution:
    
    def isPrime(self, n):
        if n <= 1:
            return False
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return False
        return True
    
    def canBeExpressed(self, n):
        for i in range(2, n // 2 + 1):
            if self.isPrime(i) and self.isPrime(n - i):
                return True
        return False


n = 74
s1 = Solution()

if s1.canBeExpressed(n):
    print("Yes")
else:
    print("No")
```

# Program to Find Roots of a quadratic equation

```python
import math
class Solution:
    def Roots(self, a, b, c):
        # Calculate discriminant
        d = b * b - 4 * a * c
        sqrt_val = math.sqrt(abs(d))

        # Case when roots are real and different
        if d > 0:
            print("Roots are real and different")
            root1 = (-b + sqrt_val) / (2 * a)
            root2 = (-b - sqrt_val) / (2 * a)
            print(root1)
            print(root2)

        # Case when roots are real and same
        elif d == 0:
            print("Roots are real and same")
            root = -b / (2 * a)
            print(root)
            print(root)

        # Case when roots are complex
        else:
            print("Roots are complex")
            realPart = -b / (2 * a)
            print(f"{realPart} + i{sqrt_val}")
            print(f"{realPart} - i{sqrt_val}")

def main():
    a, b, c = 1, -3, -10
    obj = Solution()
    obj.Roots(a, b, c)

main()
```

# Convert Decimal to Binary Number

```python
n = 54

binary = [0] * 32 
i = 0

while n > 0:
    binary[i] = n % 2
    i += 1
    n //= 2

# Print the binary number in correct order
for ind in range(i - 1, -1, -1):
    print(binary[ind], end='')
```

```python
def print_binary(n):
    flag = False  
    # Loop from 32 down to 0
    for i in range(32, -1, -1):
        # Shift n right by i bits and check the LSB
        if (n >> i) & 1:
            flag = True  # First '1' encountered
            print(1, end='')
        else:
            if flag:
                print(0, end='')

print_binary(54)
```

# Convert Decimal to Octal

# Convert Octal to Binary

```python
class Solution:
    # Convert Octal to Decimal
    def octal_to_decimal(self, Octal):
        Decimal = 0
        i = 0
        while Octal != 0:
            rem = Octal % 10
            Decimal += rem * (8 ** i)
            i += 1
            Octal //= 10
        return Decimal

    # Convert Decimal to Binary
    def decimal_to_binary(self, decimal):
        Binary = 0
        i = 0
        while decimal != 0:
            rem = decimal % 2
            Binary += rem * (10 ** i)
            i += 1
            decimal //= 2
        return Binary

if __name__ == "__main__":
    # Create solution object
    sol = Solution()
    # Example input
    Octal = 345
    # Convert octal to decimal
    Decimal = sol.octal_to_decimal(Octal)
    # Print binary conversion
    print(sol.decimal_to_binary(Decimal))
```

# Convert digits/numbers to words

```python
class Solution:
    def convert_num_into_word(self, s):
        
        single_digit = ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
        
        two_digits = ["ten", "eleven", "twelve", "thirteen", "fourteen", "fifteen", "sixteen", "seventeen", "eighteen", "nineteen"]

        tens_multiple = ["", "", "twenty", "thirty", "forty", "fifty", "sixty", "seventy", "eighty", "ninety"]

        # Words for higher powers
        tens_power = ["hundred", "thousand"]

        # Handle empty input
        if len(s) == 0:
            print("")
            return

        # Handle single digit input
        elif len(s) == 1:
            print(single_digit[int(s[0])])
            return

        # Store length of string
        length = len(s)

        # Loop through each digit
        for i in range(len(s)):
            # If more than two digits remain
            if length > 2:
                # Print digit and its place value if digit is not zero
                if int(s[i]) != 0:
                    print(single_digit[int(s[i])], end=" ")
                    print(tens_power[length - 3], end=" ")
                length -= 1
            else:
                # Handle numbers between 10 and 19
                if int(s[i]) == 1:
                    print(two_digits[int(s[i + 1])], end=" ")
                    return
                # Handle multiples of 10 and following digit
                elif int(s[i]) != 0:
                    print(tens_multiple[int(s[i])], end=" ")
                    if int(s[i + 1]) != 0:
                        print(single_digit[int(s[i + 1])], end=" ")
                    return

if __name__ == "__main__":
    # Create solution object
    sol = Solution()

    # Example input
    num = "9090"

    # Call method
    sol.convert_num_into_word(num)
```
