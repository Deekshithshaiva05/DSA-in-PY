# GRAPH

### Graph Representation

A Graph in Data Structures and Algorithms (DSA) is a non-linear data structure used to represent connections or relationships between objects.

A graph consists of:

    -> Vertices (Nodes) → the points or objects
    -> Edges → the connections between the nodes

##### 1. Adjacency Matrix (Undirected Graph)

```python
# Number of nodes and edges
n, m = map(int, input().split())

# Create adjacency matrix
adj = [[0] * (n + 1) for _ in range(n + 1)]

# Input edges
for _ in range(m):
    u, v = map(int, input().split())

    adj[u][v] = 1
    adj[v][u] = 1   # Remove this line for directed graph

# Print adjacency matrix
for row in adj:
    print(row)
```

##### 2. Adjacency List (Undirected Graph)

```python
# Number of nodes and edges
n, m = map(int, input().split())

# Create adjacency list
adj = [[] for _ in range(n + 1)]

# Input edges
for _ in range(m):
    u, v = map(int, input().split())

    adj[u].append(v)
    adj[v].append(u)   # Remove this line for directed graph

# Print adjacency list
for i in range(1, n + 1):
    print(i, "->", adj[i])
```

##### 3. Directed Graph (Adjacency List)

```python
n, m = map(int, input().split())

adj = [[] for _ in range(n + 1)]

for _ in range(m):
    u, v = map(int, input().split())

    adj[u].append(v)

for i in range(1, n + 1):
    print(i, "->", adj[i])
```

#### 4. Weighted Graph Representation

```python
n, m = map(int, input().split())

adj = [[] for _ in range(n + 1)]

for _ in range(m):
    u, v, w = map(int, input().split())

    adj[u].append((v, w))
    adj[v].append((u, w))   # Remove for directed graph

for i in range(1, n + 1):
    print(i, "->", adj[i])
```
